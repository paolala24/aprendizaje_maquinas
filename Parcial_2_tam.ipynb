```python
%%writefile PUNTO_A.py

# 1_Modelos_y_Optimizacion.py

import streamlit as st

st.set_page_config(page_title="Modelos y Optimizaci√≥n", page_icon="üìä")

st.title("üìä Modelos y Problemas de Optimizaci√≥n")

# ======================
# PCA
# ======================
st.header("1. PCA (An√°lisis de Componentes Principales)")

st.markdown("""
- **Tipo**: No supervisado, reducci√≥n de dimensionalidad
- **Prop√≥sito**: Encontrar nuevas variables (componentes) lineales que expliquen la mayor varianza posible
- **Supuestos**: Las direcciones con mayor varianza contienen m√°s informaci√≥n √∫til

#### üîÑ Proceso:
1. Centrar los datos:
   $$
   X = X - \\mu
   $$
2. Calcular matriz de covarianza:
   $$
   \\Sigma = \\frac{1}{n} X^\\top X
   $$
3. Buscar direcci√≥n $ w $ que maximice la varianza:
   $$
   \\text{Var}(Xw) = w^\\top \\Sigma w \\quad \\text{con } \\|w\\|^2 = 1
   $$

#### ‚úÖ Problema de optimizaci√≥n:
$$
\\boxed{\\max_{w} \\quad w^\\top \\Sigma w \\quad \\text{s.a. } \\|w\\|^2 = 1}
$$
La soluci√≥n se obtiene con los autovectores de $\\Sigma$.
""")

# ======================
# UMAP
# ======================
st.header("2. UMAP (Uniform Manifold Approximation and Projection)")

st.markdown("""
- **Tipo**: No supervisado, reducci√≥n de dimensionalidad no lineal
- **Objetivo**: Preservar la estructura local
- **Supuestos**: Datos residen en una variedad Riemanniana

#### üîÑ Proceso:
1. Construcci√≥n de grafo local
2. C√°lculo de probabilidades $p_{ij}$ con kernels
3. Definir distribuci√≥n $q_{ij}$ en el espacio reducido
4. Minimizar la p√©rdida de divergencia cruzada

#### ‚úÖ Problema de optimizaci√≥n:
$$
\\boxed{
\\mathcal{L} = \\sum_{i < j} \\left[ -p_{ij} \\log(q_{ij}) - (1 - p_{ij}) \\log(1 - q_{ij}) \\right]
}
$$
""")

# ======================
# Gaussian Naive Bayes
# ======================
st.header("3. Gaussian Naive Bayes")

st.markdown("""
- **Tipo**: Clasificador supervisado probabil√≠stico
- **Supuestos**: Independencia entre caracter√≠sticas dado la clase, cada caracter√≠stica es Gaussiana

#### üîÑ Proceso:
1. Teorema de Bayes:
   $$
   P(y=c | x) = \\frac{P(x | y=c) P(y=c)}{P(x)}
   $$
2. Independencia condicional:
   $$
   P(x | y=c) = \\prod_j P(x_j | y=c)
   $$
3. Gaussiana por caracter√≠stica:
   $$
   P(x_j | y=c) = \\mathcal{N}(x_j; \\mu_{jc}, \\sigma_{jc}^2)
   $$

#### ‚úÖ No requiere optimizaci√≥n iterativa:
Se estiman directamente:
$\\mu_{jc}, \\sigma_{jc}^2, P(y=c)$
""")

# ======================
# SGDClassifier
# ======================
st.header("4. SGDClassifier")

st.markdown("""
- **Tipo**: Clasificador supervisado lineal
- **Objetivo**: Minimizar p√©rdida con gradiente estoc√°stico
- **Supuestos**: Funci√≥n de p√©rdida diferenciable

#### üîÑ Proceso:
1. Elegir p√©rdida (e.g. hinge o log-loss)
2. A√±adir regularizaci√≥n $R(w)$
3. Actualizar con SGD

#### ‚úÖ Problema de optimizaci√≥n:
$$
\\boxed{
\\min_w \\frac{1}{n} \\sum_{i=1}^n \\mathcal{L}(y_i, w^\\top x_i) + \\lambda R(w)
}
$$
""")

# ======================
# Logistic Regression
# ======================
st.header("5. Logistic Regression")

st.markdown("""
- **Tipo**: Clasificador probabil√≠stico supervisado
- **Supuestos**: Separabilidad lineal

#### üîÑ Proceso:
1. Funci√≥n sigmoide:
   $$
   P(y=1|x) = \\frac{1}{1 + e^{-w^\\top x}}
   $$
2. P√©rdida logar√≠tmica:
   $$
   \\mathcal{L}(y, \\hat{y}) = -y \\log(\\hat{y}) - (1 - y)\\log(1 - \\hat{y})
   $$
3. Regularizaci√≥n L2

#### ‚úÖ Problema de optimizaci√≥n:
$$
\\boxed{
\\min_{w} \\frac{1}{n} \\sum_{i=1}^{n} \\log(1 + e^{-y_i w^\\top x_i}) + \\lambda \\|w\\|^2
}
$$
""")

# ======================
# LDA
# ======================
st.header("6. Linear Discriminant Analysis (LDA)")

st.markdown("""
- **Tipo**: Clasificador supervisado
- **Supuestos**: Clases con covarianza com√∫n

#### üîÑ Proceso:
1. Calcular medias y matrices $S_W$, $S_B$
2. Maximizar raz√≥n de varianza proyectada

#### ‚úÖ Problema de optimizaci√≥n:
$$
\\boxed{
\\max_w \\frac{w^\\top S_B w}{w^\\top S_W w}
}
$$
""")

# ======================
# KNN
# ======================
st.header("7. K-Nearest Neighbors (KNN)")

st.markdown("""
- **Tipo**: Clasificador no param√©trico
- **Supuestos**: Las muestras cercanas comparten clase

#### üîÑ Proceso:
1. Calcular distancias
2. Tomar $k$ vecinos m√°s cercanos
3. Votar por mayor√≠a

#### ‚úÖ No hay entrenamiento:
$$
\\boxed{
\\hat{y} = \\text{majority\\_vote}(y_{\\text{vecinos\ m√°s\ cercanos}})
}
$$
""")

# ======================
# SVC
# ======================
st.header("8. SVC (Support Vector Classifier)")

st.markdown("""
- **Tipo**: Clasificador de margen m√°ximo
- **Supuestos**: Separabilidad con margen (lineal o kernel)

#### üîÑ Proceso:
1. Restricciones:
   $$
   y_i(w^\\top x_i + b) \\geq 1 - \\xi_i
   $$
2. Penalizaci√≥n de errores

#### ‚úÖ Problema de optimizaci√≥n:
$$
\\boxed{
\\min_{w, b, \\xi} \\frac{1}{2}\\|w\\|^2 + C \\sum \\xi_i \\quad \\text{s.a. } y_i(w^\\top x_i + b) \\geq 1 - \\xi_i
}
$$
""")

# ======================
# Random Forest
# ======================
st.header("9. Random Forest Classifier")

st.markdown("""
- **Tipo**: Ensamble de √°rboles supervisado
- **Supuestos**: Submuestreo mejora generalizaci√≥n

#### üîÑ Proceso:
1. Entrenamiento con bootstrap
2. Divisi√≥n por impureza (Gini/Entrop√≠a)

#### ‚úÖ Optimizaci√≥n local:
$$
\\boxed{
\\min_{\\text{split}} \\; \\text{Gini o Entrop√≠a}
}
$$
Predicci√≥n por votaci√≥n:
$$
\\boxed{
\\hat{y} = \\text{majority\\_vote}(h_j(x))
}
$$
""")

# ======================
# Gaussian Process
# ======================
st.header("10. Gaussian Process Classifier")

st.markdown("""
- **Tipo**: Clasificador no param√©trico
- **Supuestos**: Funciones siguen un proceso Gaussiano

#### üîÑ Proceso:
1. Definir kernel y media
2. Calcular posterior
3. Aproximar integral marginal

#### ‚úÖ Problema de optimizaci√≥n:
$$
\\boxed{
\\log p(y \\mid X) \\text{ (aproximado con Laplace, EP o VI)}
}
$$
""")

# ======================
# Deep Learning
# ======================
st.header("11. Deep Learning Classifiers")

st.markdown("""
- **Tipo**: Clasificador con redes neuronales
- **Supuestos**: Composici√≥n de funciones no lineales

#### üîÑ Proceso:
1. Capas ocultas:
   $$
   a^{(l)} = \\phi(W^{(l)} a^{(l-1)} + b^{(l)})
   $$
2. Salida softmax:
   $$
   \\hat{y} = \\text{softmax}(a^{(L)})
   $$
3. P√©rdida de entrop√≠a cruzada:
   $$
   \\mathcal{L}(y, \\hat{y}) = - \\sum_k y_k \\log(\\hat{y}_k)
   $$

#### ‚úÖ Problema de optimizaci√≥n:
$$
\\boxed{
\\min_\\theta \\frac{1}{n} \\sum \\mathcal{L}(f_\\theta(x_i), y_i) + \\lambda \\|\\theta\\|^2
}
$$
""")


```

# **P√°gina 2**

```python
%%writefile PUNTO_B.py

# 2_USPS_Proyeccion_PCA_UMAP.py

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import umap
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

# Configurar estilo
plt.style.use("seaborn-v0_8")

st.set_page_config(page_title="Proyecciones USPS", page_icon="üìå")
st.title("üìå Proyecci√≥n del dataset USPS con PCA y UMAP")
st.markdown("Visualizaci√≥n en 2D del dataset de d√≠gitos escritos a mano (USPS) usando t√©cnicas de reducci√≥n de dimensionalidad.")

# 1. Cargar dataset
@st.cache_data
def cargar_datos():
    usps = fetch_openml("usps", version=1, as_frame=False)
    X = usps.data
    y = usps.target.astype(int)
    return X, y

X, y = cargar_datos()
st.write(f"üìê **Dimensi√≥n original:** {X.shape}")

# 2. Escalamiento
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

st.subheader("üìä Proyecci√≥n con PCA")
fig1, ax1 = plt.subplots(figsize=(10, 6))
scatter = ax1.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='tab10', s=10, alpha=0.7)
legend1 = ax1.legend(*scatter.legend_elements(), title="Etiquetas", loc="best")
ax1.add_artist(legend1)
ax1.set_title("PCA - Proyecci√≥n a 2D")
ax1.set_xlabel("PC1")
ax1.set_ylabel("PC2")
ax1.grid(True)
st.pyplot(fig1)

# 4. UMAP
st.subheader("üìä Proyecci√≥n con UMAP (n_neighbors=15)")
umap_model = umap.UMAP(n_neighbors=15, random_state=42)
X_umap = umap_model.fit_transform(X_scaled)

fig2, ax2 = plt.subplots(figsize=(10, 6))
scatter2 = ax2.scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='tab10', s=10, alpha=0.7)
legend2 = ax2.legend(*scatter2.legend_elements(), title="Etiquetas", loc="best")
ax2.add_artist(legend2)
ax2.set_title("UMAP - Proyecci√≥n a 2D (n_neighbors = 15)")
ax2.set_xlabel("UMAP1")
ax2.set_ylabel("UMAP2")
ax2.grid(True)
st.pyplot(fig2)

# 5. Superposici√≥n de im√°genes
st.subheader("üñºÔ∏è Im√°genes superpuestas")

def plot_with_images(X_proj, images, labels, title, max_images=50):
    fig, ax = plt.subplots(figsize=(12, 7))
    ax.set_title(title)
    scatter = ax.scatter(X_proj[:, 0], X_proj[:, 1], c=labels, cmap='tab10', s=10, alpha=0.5)
    shown_images = 0
    for i in range(len(X_proj)):
        if shown_images >= max_images:
            break
        if np.random.rand() < 0.02:
            img = OffsetImage(images[i].reshape(16, 16), zoom=0.5, cmap='gray')
            ab = AnnotationBbox(img, (X_proj[i, 0], X_proj[i, 1]), frameon=False)
            ax.add_artist(ab)
            shown_images += 1
    ax.set_xlabel("Componente 1")
    ax.set_ylabel("Componente 2")
    ax.grid(True)
    return fig

images = X.reshape(-1, 16, 16)

with st.expander("Mostrar PCA con im√°genes"):
    fig3 = plot_with_images(X_pca, X, y, "PCA con im√°genes superpuestas")
    st.pyplot(fig3)

with st.expander("Mostrar UMAP con im√°genes"):
    fig4 = plot_with_images(X_umap, X, y, "UMAP con im√°genes superpuestas")
    st.pyplot(fig4)

# 6. Comparaci√≥n con diferentes n_neighbors
st.subheader("üìà Comparaci√≥n de UMAP con distintos valores de n_neighbors")

neighbors_list = [5, 15, 50]
fig5, axes = plt.subplots(1, 3, figsize=(21, 6))

for i, n in enumerate(neighbors_list):
    umap_model = umap.UMAP(n_neighbors=n, random_state=42)
    X_umap_n = umap_model.fit_transform(X_scaled)
    axes[i].scatter(X_umap_n[:, 0], X_umap_n[:, 1], c=y, cmap='tab10', s=10, alpha=0.7)
    axes[i].set_title(f"UMAP (n_neighbors = {n})")
    axes[i].set_xlabel("UMAP1")
    axes[i].set_ylabel("UMAP2")

plt.tight_layout()
st.pyplot(fig5)

# 7. An√°lisis final
st.subheader("üß† An√°lisis y Discusi√≥n")
st.markdown("""
- **PCA** proyecta los datos maximizando la varianza, pero no capta bien la estructura no lineal.
- **UMAP**, al preservar la estructura local, revela agrupaciones m√°s claras.
- **n_neighbors**:
    - `n=5`: m√°s fragmentado, detalles locales.
    - `n=15`: balance entre detalle local y global.
    - `n=50`: agrupaci√≥n m√°s global.

**Conclusi√≥n**: El par√°metro `n_neighbors` controla el nivel de granularidad de la proyecci√≥n en UMAP.
""")
```

```python
%%writefile PUNTO_C.py

# 3_Modelos_Clasificacion_USPS.py

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize, StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.datasets import fetch_openml
import umap
from sklearn.decomposition import PCA
import pandas as pd

# =======================
# CONFIGURACI√ìN Y DATOS
# =======================
plt.style.use("seaborn-v0_8")
st.set_page_config(page_title="Modelos de Clasificaci√≥n", page_icon="üß†")
st.title("üß† Clasificaci√≥n con USPS: PCA y UMAP")

@st.cache_data
def cargar_datos():
    usps = fetch_openml("usps", version=1, as_frame=False)
    X = usps.data
    y = usps.target.astype(int)
    return X, y

X, y = cargar_datos()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Proyecciones
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

umap_model = umap.UMAP(n_neighbors=15, random_state=42)
X_umap = umap_model.fit_transform(X_scaled)

# Split PCA y UMAP
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, stratify=y, random_state=42)
X_train_umap, X_test_umap, _, _ = train_test_split(X_umap, y, test_size=0.3, stratify=y, random_state=42)

y_test_bin = label_binarize(y_test, classes=list(range(10)))


# =======================
# FUNCI√ìN GENERAL
# =======================
def train_and_display(model, X_train, X_test, y_train, y_test, name=""):
    st.subheader(f"üîç {name}")

    # Entrenar modelo
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # === Tabla de clasificaci√≥n
    report_dict = classification_report(y_test, y_pred, output_dict=True)
    df_report = pd.DataFrame(report_dict).transpose()
    st.markdown("**üìä Reporte de Clasificaci√≥n**")
    st.dataframe(df_report.style.format("{:.2f}").background_gradient(cmap="Greens"))

    # === Matriz de Confusi√≥n
    cm = confusion_matrix(y_test, y_pred)
    fig_cm, ax_cm = plt.subplots(figsize=(7, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=ax_cm)
    ax_cm.set_title(f"Matriz de Confusi√≥n - {name}")
    ax_cm.set_xlabel("Predicci√≥n")
    ax_cm.set_ylabel("Etiqueta Real")
    st.pyplot(fig_cm)

    # === Curvas ROC si es posible
    if hasattr(model, "predict_proba"):
        y_score = model.predict_proba(X_test)
        fpr, tpr, roc_auc = {}, {}, {}
        for i in range(10):
            fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])

        fig_roc, ax_roc = plt.subplots(figsize=(10, 7))
        for i in range(10):
            ax_roc.plot(fpr[i], tpr[i], label=f"Clase {i} (AUC = {roc_auc[i]:.2f})")
        ax_roc.plot([0, 1], [0, 1], "k--")
        ax_roc.set_xlabel("Tasa de Falsos Positivos (FPR)")
        ax_roc.set_ylabel("Tasa de Verdaderos Positivos (TPR)")
        ax_roc.set_title(f"Curvas ROC - {name}")
        ax_roc.legend(fontsize=8)
        ax_roc.grid(True)
        st.pyplot(fig_roc)


# =======================
# CLASIFICADORES
# =======================
with st.expander("1Ô∏è‚É£ Logistic Regression (PCA)", expanded=True):
    lr_pca = LogisticRegression(max_iter=500, solver="lbfgs", multi_class="multinomial")
    train_and_display(lr_pca, X_train_pca, X_test_pca, y_train, y_test, "Logistic Regression (PCA)")

with st.expander("2Ô∏è‚É£ Random Forest (UMAP)", expanded=True):
    rf_umap = RandomForestClassifier(n_estimators=100, max_depth=15, random_state=42)
    train_and_display(rf_umap, X_train_umap, X_test_umap, y_train, y_test, "Random Forest (UMAP)")

with st.expander("3Ô∏è‚É£ MLP Neural Net (PCA)", expanded=True):
    mlp_pca = make_pipeline(
        StandardScaler(),
        MLPClassifier(hidden_layer_sizes=(128, 64), activation="relu",
                      max_iter=200, alpha=1e-4, random_state=42)
    )
    train_and_display(mlp_pca, X_train_pca, X_test_pca, y_train, y_test, "MLP Neural Net (PCA)")


# =======================
# DISCUSI√ìN
# =======================
st.subheader("üìå An√°lisis Final")
st.markdown("""
- **Logistic Regression (PCA)**: Buen rendimiento si las clases son linealmente separables. R√°pido, pero limitado.
- **Random Forest (UMAP)**: Ideal para datos con estructura compleja. Captura relaciones no lineales.
- **MLPClassifier (PCA)**: Alta capacidad para aprender patrones complejos. Mejora con escalamiento.

**Curvas ROC** permiten evaluar el desempe√±o clase por clase. Una curva m√°s cerca de la esquina superior izquierda indica mejor desempe√±o.
""")


```
